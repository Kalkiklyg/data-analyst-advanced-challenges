```python
# Day 6 — Visualizing Insights for Business Clarity

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1️⃣ Sample sales dataset
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Revenue': [12000, 18000, 17000, 24000, 26000, 23000],
    'Expenses': [8000, 12000, 11000, 15000, 14000, 16000]
}
df = pd.DataFrame(data)
df['Profit'] = df['Revenue'] - df['Expenses']

# 2️⃣ Basic line chart
plt.figure(figsize=(8,5))
sns.lineplot(x='Month', y='Revenue', data=df, marker='o', label='Revenue')
sns.lineplot(x='Month', y='Expenses', data=df, marker='o', label='Expenses')
sns.lineplot(x='Month', y='Profit', data=df, marker='o', label='Profit')
plt.title("Monthly Financial Performance")
plt.xlabel("Month")
plt.ylabel("Amount (₹)")
plt.legend()
plt.tight_layout()
plt.show()

# 3️⃣ Highlight the best month by profit
best_month = df.loc[df['Profit'].idxmax()]
print(f"Best Month: {best_month['Month']} — Profit ₹{best_month['Profit']}")

```
