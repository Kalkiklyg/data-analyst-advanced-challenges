```python
# Day 11 — Correlation & Root-Cause Analysis
# Author: Shailesh Pawar (Kalkiklyg)

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 1️⃣ Create dataset
np.random.seed(42)
data = {
    "Marketing_Spend": np.random.randint(50, 150, 12),
    "Revenue": [x + np.random.randint(-10,10) + 100 for x in np.random.randint(100,200,12)],
    "Customer_Satisfaction": np.random.uniform(3,5,12).round(2)
}
df = pd.DataFrame(data)

# 2️⃣ Compute correlation matrix
corr_matrix = df.corr(numeric_only=True)
print(corr_matrix)

# 3️⃣ Visualize
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# 4️⃣ Key insights
strong_corr = corr_matrix['Revenue'].sort_values(ascending=False)
print("\nTop factors correlated with Revenue:\n", strong_corr)

```
