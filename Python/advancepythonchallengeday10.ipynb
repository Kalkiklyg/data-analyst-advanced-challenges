```python
# Day 10 — Anomaly Detection: Spotting the Outliers Before They Hurt
# Author: Shailesh Pawar (Kalkiklyg)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1️⃣ Create a dataset (Monthly Transactions)
data = {
    "Month": ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"],
    "Transactions": [420, 450, 460, 490, 505, 495, 1000, 520, 510, 530, 545, 555]
}
df = pd.DataFrame(data)

# 2️⃣ Compute z-scores for anomaly detection
df["z_score"] = (df["Transactions"] - df["Transactions"].mean()) / df["Transactions"].std()

# 3️⃣ Flag anomalies (Z > 2 or Z < -2)
df["Anomaly"] = df["z_score"].apply(lambda x: "Yes" if abs(x) > 2 else "No")

# 4️⃣ Plot
plt.figure(figsize=(8,4))
plt.plot(df["Month"], df["Transactions"], marker='o')
plt.scatter(df.loc[df["Anomaly"]=="Yes","Month"], df.loc[df["Anomaly"]=="Yes","Transactions"],
            color='red', label="Anomaly", s=100)
plt.title("Monthly Transactions with Anomalies")
plt.xlabel("Month")
plt.ylabel("Transactions")
plt.legend()
plt.grid(True)
plt.show()


```
