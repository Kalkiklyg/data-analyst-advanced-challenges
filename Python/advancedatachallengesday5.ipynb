```python
# Day 5 — Data Transformation & Insight Extraction
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1️⃣ Load Data
data = {
    "Region": ["North", "South", "East", "West", "North", "South", "East", "West"],
    "Salesperson": ["A", "B", "C", "D", "E", "F", "G", "H"],
    "Revenue": [12000, 18000, 15000, 21000, 16000, 22000, 20000, 25000],
    "Target": [10000, 17000, 16000, 19000, 15000, 20000, 19000, 24000]
}
df = pd.DataFrame(data)

# 2️⃣ Add performance columns
df["Performance_%"] = round(df["Revenue"] / df["Target"] * 100, 2)
df["Status"] = np.where(df["Performance_%"] >= 100, "Achieved", "Pending")

# 3️⃣ Pivot table – average performance by region
pivot = df.pivot_table(values="Performance_%", index="Region", aggfunc="mean")
print("Average Performance by Region:\n", pivot)

# 4️⃣ Visualization
plt.bar(pivot.index, pivot["Performance_%"])
plt.title("Regional Performance vs Target")
plt.xlabel("Region")
plt.ylabel("Average Performance (%)")
plt.show()

```
